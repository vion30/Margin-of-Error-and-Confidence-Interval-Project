```python
# # Q1. Problem Statement: Inferential Statistics
# # Load the wine dataset using the Sklearn dataset and convert it to DataFrame perform the following task: 
# 1. Covert the wine data into DataFrame using pandas 
# 2. Generate the sample size of 50 and give a random state as 100 
# 3. Calculate Z-critical, Margin of Error & Confidence Interval for alcohol at 95% significance interval on generated sample data
```

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
wine
```

```python
df = pd.DataFrame(wine.data,columns= wine['feature_names'])
```

```python
df
```

```python
# 2. Generate the sample size of 50 and give a random state as 100 

wine = df.sample(50, random_state = 100)
wine.head()
```

```python
# 3. Calculate Z-critical, Margin of Error & Confidence Interval for alcohol at 95% significance interval on generated sample data

sample_mean = wine['alcohol'].mean()
print("sample_mean : " , sample_mean)

np.random.seed(1)

z_critical = stats.norm.ppf(q=0.95)
print("z_critical :", z_critical)

sam_st = wine['alcohol'].std()
print("sam_st : ", sam_st)

margin_error = z_critical*(sam_st/np.sqrt(50))
print("Margin of Error :", margin_error)

CI = sample_mean + (z_critical*(sam_st/np.sqrt(50))), sample_mean - (z_critical*(sam_st/np.sqrt(50)))
print("Confidence Interval :", CI)
```
